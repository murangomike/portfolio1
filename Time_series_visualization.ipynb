USA COVID_19 TEST CASES TIME SERIES VISUALIZATION

```python
import pandas as pd
import matplotlib.pyplot as plt



# Extract date columns (excluding 'country_region', 'lat', 'long', and 'location')
date_columns = [col for col in covid_test_cases.columns if col not in ['country_region', 'lat', 'long', 'location']]

# Create a new DataFrame with dates as rows
df_melted = covid_test_cases.melt(id_vars=['country_region'], value_vars=date_columns,
                                  var_name='date', value_name='cases')

# Convert 'date' to datetime format AFTER melting
df_melted['date'] = pd.to_datetime(df_melted['date'], format='%m_%d_%y')

# Set the date column as the index
df_melted.set_index('date', inplace=True)

# Plot the time series for a specific country/region
country = 'US'
df_country = df_melted[df_melted['country_region'] == country]

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df_country.index, df_country['cases'], marker='o', linestyle='-', color='b')
plt.title(f'COVID-19 Cases Over Time for {country}')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Continuous Increase: The overall trend indicates a continuous increase in the total number of cases, suggesting that while the rate of new cases might fluctuate, the cumulative count keeps rising.


Epidemic Waves: The distinct periods of rapid growth followed by slower growth or plateaus indicate multiple waves of the pandemic, reflecting changes in transmission dynamics, public health responses, and possibly the emergence of new variants.

INTERATIVE VISUALIZATION OF GLOBAL COVID-19 TEST CASES

```python
import pandas as pd
import plotly.express as px

# Assuming 'covid_test_cases' is your DataFrame

# Group by country and sum the cases across all dates
covid_test_cases_grouped = covid_test_cases.groupby('country_region').sum().reset_index()

# Convert date columns to numeric, handling non-numeric values
date_columns = covid_test_cases_grouped.columns[4:]  # Select columns starting from the 5th (index 4)
covid_test_cases_grouped[date_columns] = covid_test_cases_grouped[date_columns].apply(pd.to_numeric, errors='coerce')

# Create the geomap
fig = px.scatter_geo(covid_test_cases_grouped,
                     lat='lat',
                     lon='long',
                     size=covid_test_cases_grouped.iloc[:, 4:].sum(axis=1), # Sum cases across all date columns
                     color=covid_test_cases_grouped.iloc[:, 4:].sum(axis=1),
                     hover_name='country_region',
                     projection="natural earth")

fig.update_layout(title='Global Distribution of COVID-19 Test Cases')
fig.show()
```

GLOBAL VISUALIZATION OF TOTAL DEATHS PER COUNTRY

```python
import pandas as pd
import plotly.express as px

# Group by country and sum the death cases across all dates
covid_death_cases_grouped = covid_deaths.groupby('country_region').sum().reset_index()

# Convert date columns to numeric, handling non-numeric values
date_columns = covid_death_cases_grouped.columns[4:]
covid_death_cases_grouped[date_columns] = covid_death_cases_grouped[date_columns].apply(pd.to_numeric, errors='coerce')

# Sum death cases across all date columns
total_deaths = covid_death_cases_grouped[date_columns].sum(axis=1)

# Create the geomap
fig = px.scatter_geo(covid_death_cases_grouped,
                     lat='lat',
                     lon='long',
                     size=total_deaths,  # Sum deaths across all date columns
                     color=total_deaths,
                     hover_name='country_region',
                     projection="natural earth")

fig.update_layout(title='Global Distribution of COVID-19 Death Cases')
fig.show()
```

GLOBAL VISUALIZATION OF TOTAL RECOVERED CASES

```python
import pandas as pd
import plotly.express as px

# Group by country and sum the recovered cases across all dates
recovered_cases_grouped = recovered_cases.groupby('country_region').sum().reset_index()

# Convert date columns to numeric, handling non-numeric values
date_columns = recovered_cases_grouped.columns[4:]
recovered_cases_grouped[date_columns] = recovered_cases_grouped[date_columns].apply(pd.to_numeric, errors='coerce')

# Sum recovered cases across all date columns
total_recovered = recovered_cases_grouped[date_columns].sum(axis=1)

# Create the geomap
fig = px.scatter_geo(recovered_cases_grouped,
                     lat='lat',
                     lon='long',
                     size=total_recovered,
                     color=total_recovered,
                     hover_name='country_region',
                     projection="natural earth")

fig.update_layout(title='Global Distribution of COVID-19 Recovered Cases')
fig.show()
```

TIME SERIES VISUALIZATION OF USA, RUSSIA, BRAZIL AND INDIA.

```python
import pandas as pd
import plotly.express as px

# Assuming covid_test_cases is already loaded and melted
df_long = covid_test_cases.melt(id_vars=['country_region', 'lat', 'long', 'location'],
                                var_name='date',
                                value_name='cases')

# Convert 'date' to datetime
df_long['date'] = pd.to_datetime(df_long['date'], format='%m_%d_%y')

# Filter data for Russia and the US
countries_to_plot = ['Russia', 'US', 'UK', 'India', 'Brazil']
df_filtered = df_long[df_long['country_region'].isin(countries_to_plot)]

# Set 'date' as index
df_filtered.set_index('date', inplace=True)

# Create a line plot for the time series data
fig = px.line(df_filtered,
              x=df_filtered.index,
              y='cases',
              color='country_region',
              title='COVID-19 Time Series for Russia and the US',
              labels={'cases': 'Number of Cases', 'country_region': 'Country'})

fig.update_layout(xaxis_title='Date', yaxis_title='Number of Cases')
fig.show()
```
