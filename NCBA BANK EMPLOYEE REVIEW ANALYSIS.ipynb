<h1 style="color: white;">NCBA Bank Group Glassdoor Reviews Dataset</h1>


This dataset contains reviews from employees working at NCBA Bank Group in Kenya, sourced from Glassdoor. Each row in the dataset represents an individual employee's feedback on their work experience. Below is an explanation of each column in the dataset:

**Column Descriptions:**

Review Title:
The headline or summary of the employee's review, providing a quick overview of their experience at NCBA Bank.

Date:
The date when the employee posted the review. This helps understand how recent the feedback is.

Rating:
The overall rating given by the employee, typically on a scale from 1 to 5, with 1 being the worst experience and 5 being the best. This shows how satisfied or dissatisfied the employee is.

Position:
The job title or role of the employee posting the review. This helps identify which department or level of seniority the feedback is coming from.

Location:
The city or region where the employee works. This is important as experiences may differ depending on the location.

Pros:
The positive aspects of working at NCBA Bank that the employee has mentioned. This can include things like a good work environment, supportive management, or benefits.

Cons:
The challenges or negative aspects of the job according to the employee. This may include issues like lack of promotion opportunities, poor work-life balance, or dissatisfaction with management.

```python
from sklearn.impute import SimpleImputer
def impute(df_glassdoor, Location):
    
    # Create a SimpleImputer to fill missing values with the most frequent value (mode)
    imputer = SimpleImputer(strategy='most_frequent')
    
    # Apply the imputer on the Location column
    df_glassdoor['Location'] = imputer.fit_transform(df_glassdoor[['Location']]).ravel()
    
    # Verify if null values have been replaced
    print ("-" * 50)
    print (f" The null values in the location column are {df_glassdoor['Location'].isnull().sum()}")
    return df_glassdoor
```

## EMPLOYEE RETENTION USING RATINGS AND SENTIMENTS

**SPLIT AND TRAIN THE DATASET**

```python
# Set a threshold to determine retention (this is arbitrary, modify as needed)
retention_threshold = 3

# Assume employees with a rating >= threshold are retained
df_glassdoor['Retention'] = df_glassdoor['Rating'].apply(lambda x: 1 if x >= retention_threshold else 0)

# Calculate retention rate
retention_rate = df_glassdoor['Retention'].mean() * 100  # Convert to percentage

print(f"Employee Retention Rate: {retention_rate:.2f}%")
```

```python
# Define retention based on sentiment scores
# For example, if both Pros and Cons sentiment are positive, assume retention
df_glassdoor['Sentiment_Retention'] = df_glassdoor.apply(
    lambda row: 1 if row['Pros_Sentiment'] >= 0 and row['Cons_Sentiment'] >= 0 else 0, axis=1
)

# Calculate retention rate based on sentiment
sentiment_retention_rate = df_glassdoor['Sentiment_Retention'].mean() * 100  # Convert to percentage

print(f"Employee Retention Rate based on sentiment: {sentiment_retention_rate:.2f}%")
```

Rating-based retention (94.09%): A high retention rate, meaning most employees rate the company reasonably well (above the threshold of 3). However, ratings alone may not fully reflect their detailed thoughts.

Sentiment-based retention (65.02%): The sentiment scores suggest that a significant proportion of employees have mixed feelings about the company, as only about 65% have overall positive sentiment in both Pros and Cons. This could mean that despite acceptable ratings, many employees express dissatisfaction in specific areas.

This difference between rating and sentiment-based retention rates could indicate that while employees may feel compelled to give good ratings (for various reasons like loyalty or wanting to keep their jobs), they may still harbor negative feelings about their work environment or certain company policies.

## EMPLOYEE RETENTION(sentiments) USING LOGISTIC REGRESSION

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Creating the feature matrix and target vector
X = df_glassdoor[['Rating', 'Pros_Sentiment', 'Cons_Sentiment']]
y = df_glassdoor['Sentiment_Retention']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Logistic Regression model
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# Predict and evaluate
y_pred = logreg.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import cross_val_score

# Balance the classes using SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Train-test split on the resampled data
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# Logistic Regression model
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# Cross-validation
cv_scores = cross_val_score(logreg, X_resampled, y_resampled, cv=5)
print(f"Cross-validated accuracy: {cv_scores.mean()}")

# Predict and evaluate
y_pred = logreg.predict(X_test)
print(classification_report(y_test, y_pred))
```
