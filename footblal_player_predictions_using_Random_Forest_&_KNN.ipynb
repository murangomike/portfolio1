```python
import numpy as np
import pandas as pd


df= pd.read_csv ('football_players.csv', encoding = 'latin-1')
```

```python
#What is the most common Overall score for players in the database?

common_overall_rating = df['Overall'].mode()
print(common_overall_rating)
```

**DATA PREPROCESSING**

```python
cb =df[df['Preferred Positions'].str.contains('CB')]
cb ['Group'] = cb['Overall'].apply(lambda x: 'World class' if x >= 80 else ('Good' if 70 <= x <= 79 else 'Mediocre'))
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
cb['Preferred Positions'] = le.fit_transform(cb['Preferred Positions'])
cb['Group'] = le.fit_transform(cb['Group'])
cb['Preferred Positions Type'] = le.fit_transform(cb['Preferred Positions Type'])
cb['Nationality'] = le.fit_transform(cb['Nationality'])
```

**`DATA SPLITTING AND MODELLING WITH RANDOM FOREST**

```python
X = cb.drop(['Group'], axis=1)
y = cb['Group']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Drop the 'Name' column from X before splitting into train and test sets.
X = cb.drop(['Group', 'Name'], axis=1)
y = cb['Group']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1971)

rf = RandomForestClassifier(n_estimators=500, random_state=1971)
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)
coeff = rf.feature_importances_
coeff = pd.Series(coeff, index=X.columns)
coeff.sort_values(ascending=False)
```

MODEL EVALUATION

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
confusion_mat = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix:\n{confusion_mat}")
```

```python
classification_report = classification_report(y_test, y_pred)
print(f"Classification Report:\n{classification_report}")
```

PREDICTION WITH K-NEAREST NEIGHBOURS MODEL

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=911)

for n in range (1,6):
  print (f"n = {n}")
  knn = KNeighborsClassifier(n_neighbors=n)
  knn.fit(X_train, y_train)

  y_pred_knn = knn.predict(X_test)

  accuracy = accuracy_score(y_test, y_pred_knn)
  print(f"Accuracy: {accuracy}")
  f1_score_macro = f1_score(y_test, y_pred_knn, average='macro')
  print(f"f1 score (macro): {f1_score_macro}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
from sklearn.model_selection import train_test_split

# Assuming X and y are already defined as in your previous code

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=911)

best_f1_worldclass = 0
best_k_worldclass = 0
best_f1_good = 0
best_k_good = 0

for n in range(1, 6):
    print(f"n = {n}")
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_train, y_train)

    y_pred_knn = knn.predict(X_test)

    # Calculate F1-score for each class separately
    f1_scores = f1_score(y_test, y_pred_knn, average=None)

    # Assuming 'World Class' is the first class (index 0) and 'Good' is the second (index 1)
    f1_worldclass = f1_scores[0]
    f1_good = f1_scores[1]

    print(f"F1-score (World Class): {f1_worldclass}")
    print(f"F1-score (Good): {f1_good}")

    if f1_worldclass > best_f1_worldclass:
        best_f1_worldclass = f1_worldclass
        best_k_worldclass = n

    if f1_good > best_f1_good:
        best_f1_good = f1_good
        best_k_good = n

print("\nBest Results:")
print(f"Best k for World Class: {best_k_worldclass} with F1-score: {best_f1_worldclass}")
print(f"Best k for Good: {best_k_good} with F1-score: {best_f1_good}")
```
